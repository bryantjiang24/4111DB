#### Questions 1: DML and DDL

- Briefly explain the concepts of data definition language and data manipulation language.
- Give one example of a SQL DDL statement and a SQL DML statement.

_Answer:_ <br>
DDL is Data Definition Language which is used to define data structures. In SQL, for example: create table, alter table.<br> 
DML is Data Manipulation Language which is used to manipulate data itself. In SQL, for example: insert, update, delete.

#### Questions 3: Types of Data

- Briefly explain the concepts of structured, semi-structured and unstructured data.
- Give an example of each type of data.

_Answer:_ <br>
Structured data are addressable for effective analysis, which has been organized into a database. It concerns all data which can be stored in a table with rows and columns. They have relational keys and can easily be mapped into pre-designed fields. Example: Relational data. <br>
Semi-structured data is information which not resides in a relational database but has some organizational properties to analyze. You can store them in the relation database with some processes, but it exist to ease space. Example: XML data. <br>
Unstructured data is not organized in a predefined manner or does not have a predefined data model. It is not a good fit for a mainstream relational database. Example: Word, PDF, Text. 

#### Questions 4: Data Abstraction Levels

- Briefly explain the concepts of physical level, logical level and view level in databases.


_Answer:_ <br>
Physical level tells us where the data is actually stored, the actual location of the data that is being stored by the user. <br>
Logical level tells how the data is actually stored and structured, the overall logical structure of the database. <br>
View level tells the application about how the data should be shown to the user, which is the closest level to the user.

#### Questions 5: Instance and Schema

- Would you use DML or DDL statements to create and update a schema?

_Answer:_ <br>
A database schema is designed to help programmers whose software will interact with the database. Logical schema is the overall logical structure of the database. Physical schema is the overall physical structure of the database.<br>

#### Questions 6: Declarative versus Procedural Languages

- Briefly explain the difference between a procedure data manipulation language and a procedural data manipulation language.
- Briefly explain why relational algebra is procedural and Python is declarative.

_Answer:_ <br>
Procedural DML requires a user to specify what data are needed and how to get those data.<br>
Declarative DML requires a user to specify what data are needed without specifying how to get those data.<br>
Relational algebra is procedural. In relational algebra, the order is specified in which the operations have to be performed. <br>
Python is declarative. It let you define what data you want computed and translate that efficiently onto the database schema.

#### Questions 7: Application Architectures

- Briefly explain the concepts of two-tier and three-tier application architectures.
- Are Jupyter notebooks two-tier or three-tier applications?

_Answer:_ <br>
In two-tier, the application logic is either buried inside the User Interface on the client or within the database on the server (or both). <br>
In three-tier, the application logic or process lives in the middle-tier, it is separated from the data and the user interface. Three-tier systems are more scalable, robust and flexible. In addition, they can integrate data from multiple sources. <br>
Jupyter notebooks are three-tier applications.
