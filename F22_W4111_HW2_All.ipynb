__Answer:__

1. Security. Each user can be given permission to access the database only through a small set of views that contain the specific data the user is authorized to see, thus restricting the user's access to stored data

## C3

__Question__

- Use the tables ```people, appearances, batting``` from the Lahman's Baseball data to answer this question.


- Produce a table of the form:
    - ```playerID```
    - ```nameLast```
    - ```nameFirst```
    - ```career_teams``` is a semi-colon separated list of the team
    - ```career_games``` is the sum of ```G_all``` from ```appearances```
    - ```total_abs``` is the sum of ```AB``` from ```batting```
    - ```total_hits``` is the sum of ```h``` from ```batting```
    - ```batting_avg``` is ```total_hits/total_abs``` is ```total_abs``` is not 0, and is ```NULL`` otherwise.
    

- Show the first 10 rows like below.



__Answer 1__

````python
%%sql use lahmansbaseballdb;
WITH career_teams as (
    SELECT playerID, 
    SUM(G_all) as 'career_games' 
    FROM appearances 
    GROUP BY 1), 
    batting as (
        SELECT playerID,
                SUM(AB) as total_abs,
                SUM(h) as total_hits
        FROM batting
        GROUP BY 1
    )

SELECT p.playerID,
       p.nameLast,
       p.nameFirst, 
       GROUP_CONCAT(DISTINCT a.teamID,";") as career_teams,
       c.career_games as career_games,
       b.total_abs,
       b.total_hits,
       ROUND(total_hits/ (CASE WHEN total_abs !=0 THEN total_abs ELSE  NULL END),4) as batting_avg

FROM appearances as a 
INNER JOIN people as p 
ON a.playerID = p.playerID
INNER JOIN career_teams as c 
ON c.playerID = a.playerID
INNER JOIN batting as b
ON b.playerID = a.playerID
GROUP BY a.playerID
LIMIT 10;
````

## C4

__question__


- A person (from ```people```) was a player in MLB if their ```playerID``` appears in ```appearances.```


- A person (from ```managers```) was a manager if their ```playerID``` appears in ```managers.```


- Produce the following table from ```halloffame``` for people in ```halloffame``` that were not managers or players.


- My first 10 rows look like below.





- Your query should produce all rows.

__Answers__

````python
%%sql

WITH app as (
    SELECT playerID as 'playerid'
    FROM people
    WHERE playerID in (
        SELECT playerID
        FROM appearances
    )
    UNION all 
    SELECT playerID as 'playerid'
    FROM managers
)


SELECT h.playerID as 'playerid',
       p.nameLast,
       p.nameFirst,
       h.category
FROM halloffame as h
LEFT JOIN people as p
ON p.playerID = h.playerID
WHERE h.playerID not in (
    SELECT playerid
    FROM app
)
LIMIT 10
````
